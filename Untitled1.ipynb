```python
url="https://raw.githubusercontent.com/juarezefren/datasets/main/Iris.csv"
import pandas as pd
df = pd.read_csv(url)
df
```

```python
from sklearn.utils import shuffle
# Revolver los registros
df_shuffled = shuffle(df, random_state=42)

# Reiniciar los índices del DataFrame revuelto
df_shuffled.reset_index(drop=True, inplace=True)

df_shuffled
```

```python
X_train= df_shuffled.iloc[:100,1:-1]
X_train
```

```python
Y_train= df_shuffled.iloc[:100,-1]
Y_train
```

```python
X_test = df_shuffled.iloc[100:,1:-1]
X_test
```

```python
Y_test = df_shuffled.iloc[100:,-1]
Y_test
```

```python
from sklearn.svm import SVC
modelo = SVC()
modelo.fit(X_train,Y_train)
modelo
```

```python
y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
precision = accuracy_score(Y_test, y_pred)
precision
```
