```python
url="https://raw.githubusercontent.com/juarezefren/datasets/main/insects-10.csv"
import pandas as pd
df = pd.read_csv(url)
df
```

```python
X_train= df.iloc[:,1:-1]
X_train
```

```python
Y_train= df.iloc[:,-1]
Y_train
```

```python
from sklearn.svm import SVC
modelo = SVC()
modelo.fit(X_train,Y_train)
modelo
```

```python
ejemplo = pd.DataFrame( {'abdomen':[5.1],'antena':[7.0]})
ejemplo
```

```python
modelo.predict(ejemplo)
```

```python
X_test = pd.DataFrame( {'abdomen':[5.1, 2, 2, 1.5,7],'antena':[7.0,2.0, 4.0,5.0, 8.0 ]})
X_test
```

```python
predicted = modelo.predict(X_test)
predicted
```
